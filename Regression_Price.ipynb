```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('price_of_bottle.csv')
df
```

```python
df2=df.drop(['Continent'],axis=1)
```

```python
df2_1 = pd.read_csv('regression_pop.csv')
df2_1
```

```python
df2_2 = df2_1.drop(['Consumption (litres)', 'Consumption','GDP pc'],axis=1)
df2_2
```

```python
df2_3 = df2_2.rename(columns={'Country Name': 'Country'})
```

```python
merge_df = df2.merge(df2_3, on=['Country'], how='left')
merge_df
```

```python
exported_values = list(merge_df.T.to_dict().values())
```

```python
import json
```

```python
open('regression.json','w').write(json.dumps(exported_values))
```
