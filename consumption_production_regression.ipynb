```python
import pandas as pd
```

```python
df1_1 = pd.read_csv('regression_data.csv')
df1_1
```

```python
df1_2 = df1_1.drop(['Consumption'],axis=1)
df1_2
```

```python
df1_3 = df1_2.rename(columns={'Country Name': 'Country'})
```

```python
df2_1 = pd.read_csv('country_production.csv')
df2_1
```

```python
merge_df = df1_3.merge(df2_1, on=['Country', 'Continent'], how='right')
merge_df
```

```python
merge_df['Prod_pc'] = merge_df['Production (litres)']/merge_df['Population']
merge_df
```

```python
exported_values = list(merge_df.T.to_dict().values())
```

```python
import json
```

```python
open('final_consumption.json','w').write(json.dumps(exported_values))
```
