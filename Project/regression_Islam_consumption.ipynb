```python
import pandas as pd
import requests
```

Now I am creating a dataframe that is reading the html code of the website below. This code reads the html and selects the tables from the site. By using inspect element I was able to determine that the table could be scraped in Colab. The Table shows the Muslim population in every country in a study conducted by the Pew Forum Research Centre

```python
df = pd.read_html('https://web.archive.org/web/20110202124356/http://features.pewforum.org/muslim-population/')
df
```

Having created a dataframe with the tables from the site, I am checking how many tables are in the array scraped and which one has the data desired. From using the len() function I know that there are 2 tables in the array and from looking at each one I know that the second table in the array contains the desired information

```python
len(df)
```

```python
df2 = df[1]
df2
```

Now that I have scraped the table succeffully into Colab I am dropping the columns that are unnecessary for the regression I am going to undertake. These are the Projections for 2030 and the 1990 values. I am also renaming the columns to simplify my code

```python
df3 = df2.drop(['Estimated 1990 Muslim Population','Percentage of 1990 Population that is Muslim','Projected 2030 Muslim Population*','Projected Percentage of 2030 Population that is Muslim*'],axis=1)
df3
```

```python
df4 = df3.rename(columns={'Estimated 2010 Muslim Population':'Muslim Population','Percentage of 2010 Population that is Muslim':'Percentage'})
```

Having renamed the columns I needed to rename a number of the Country names so that they would maytch up with the other dataset when merging the two. Therefore I manually sorted through the data and renamed the required countries so as to not affect the dataset

```python
df4['Country'] = df4['Country'].replace({'Swaziland': 'Eswatini','Congo':'The Democratic Republic of Congo','Cape Verde':'Cabo Verde','Guinea Bissau':'Guinea-Bissau',"Ivory Coast":"Cote d'Ivoire", 'Laos':"Lao People's Democratic Republic",'Burma (Myanmar)': 'Myanmar', 'South Korea':'Korea, Republic of', 'Faeroe Islands':'Faroe Islands','Bosnia-Herzegovina':'Bosnia and Herzegovina','Republic of Macedonia': 'North Macedonia', 'Slovakia':'Slovak Republic','Czech Republic':'Czechia','Russia':"Russian Federation",'St. Kitts and Nevis': 'Saint Kitts and Nevis','St. Vincent and the Grenadines':'Saint Vincent and the Grenadines','St. Lucia':'Saint Lucia','Macau':'Macao'})
df5=df4
```

Now that I have my cleaned data for the first source (Muslim %), I am importing a csv of Wine Consumption per capita (calculated in Excel).

```python
df2_1 = pd.read_csv('regression_data.csv')
df2_1
```

```python
df2_2 = df2_1.drop(['Consumption'],axis=1)
df2_2
```

```python
df2_3 = df2_2.rename(columns={'Country Name': 'Country'})
```

I had to rename the 'Country Name' column since the two columns need to match up when merging datasets. Below I have merged the two datasets. In the code I have merged based on Country name (hence why editing the names was so vital earlier) and i have merged it on the 'right' dataset. This is the df2_3 dataset since it has fewer values so less sorting is required after merging

```python
merge_df = df5.merge(df2_3, on=['Country'], how='right')
merge_df
```

Having done this I recognised that some of the data in the table will affect my regression. Since I am regressing percentage I need to remove all the observations that are labelled "<0.1", they are not numeric values and are nearing 0 so I have removed them at the extremety of observations

I used the drop() function to remove all the rows with that as their value for Percentage

```python
merge_df2 = merge_df.drop(merge_df[merge_df['Percentage']=='< 0.1'].index)
```

```python
exported_values = list(merge_df2.T.to_dict().values())
```
