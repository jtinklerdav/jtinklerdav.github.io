```python
import pandas as pd
```

```python
df = pd.read_csv('final_consumption.csv')
df
```

```python
df1 = df.drop(['Region'],axis=1)
df1
```

```python
df2 = pd.read_csv('pop_overtime.csv')
df2
```

```python
df3=df2.set_index(['Country']).stack().reset_index()
df3
```

```python
df4 = df3.rename(columns={'level_1': 'Year',0:'population'})
```

```python
df4['Year']=df4['Year'].astype(int)
df1['Year']=df1['Year'].astype(int)
```

```python
merge_df = df4.merge(df1, on=['Country', 'Year'], how='right')
merge_df
```

```python
merge_df['cons_pc'] = 100000* merge_df['Wine Consumption (1000hl)']/merge_df['population']
merge_df
```

```python
exported_values = list(merge_df.T.to_dict().values())
```

```python
import json
```

```python
open('final_consumption', 'w').write(json.dumps(exported_values))
```
